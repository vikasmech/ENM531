### Subproblem 1

A train bridge is constructed across a wide river. Trains arrive at the bridge according to a Poisson process of rate $\lambda = 3$ per day.

- Find the probability that no trains arrive in the first two days, but 4 trains arrive on the 4th day.
- Find the probability that it takes more than 2 days for the 5th train to arrive at the bridge.

### Answer

- $\mathbb{P}$(No trains on first two days and 4 trains on 4th day) = $\mathbb{P}$( No trains on day 1)P( No trains on day 2)P(4 trains on day 4) = $e^{-3} e^{-3} \frac{e^{-3} 3^4}{4!}$ = $\frac{e^{-9} 3^4}{4!} \approx 0.0004165$
- In a Poisson process, we have that the time of the k-th arrival has the following density
\begin{align}
f_{Y_k}(t) = \frac{\lambda e^{-\lambda t} (\lambda t)^{k - 1}}{(k - 1)!}, k = 1, 2, \cdots, t > 0
\end{align}

Hence, the time for the fifth arrival for $\lambda = 3$ has the density

\begin{align}
    f_{Y_5}(t) = \frac{3 e^{-3 t} (3 t)^{4}}{4!}
\end{align}

We are interested in
\begin{align}
    \mathbb{P}(X_5 > 2) = \int_2^{\infty} \frac{3 e^{-3 t} (3 t)^{4}}{4!} dt = 115 e^{-6} \approx 0.285
\end{align}

### Subproblem 2

Let $X$ and $Y$ have joint PDF 
$f_{X,Y}(x,y) = \begin{cases}
C e^{-(ax+by)} & x,y \geq 0\\
0 & \text{otherwise}
\end{cases},$

where, $a, b > 0$ are constants

- Determine the constant $C$
- Find the marginal density of $X$ and $Y$. What can you infer from this?
- Find $\mathbb{E}(Y \mid X> \frac{\exp(a + b)}{a^2 + b^2})$
- Find $\mathbb{P}(0 < X < \sqrt{Y})$

### Answer

- Using that $\int_{\mathbb{R}^2} f_{X, Y}(x, y) = 1$, we have $C = ab$
- We have

\begin{align}
    f_X(x) = \int_{-\infty}^{\infty} f_{X, Y} (x, y) dy = a e^{-ax} \\
    f_Y(y) = \int_{-\infty}^{\infty} f_{X, Y} (x, y) dx = b e^{-by}
\end{align}

We have that $f_{X, Y} (x, y) = f_X \cdot f_Y$. Hence, $X$ and $Y$ are independent.
- Since $X$ and $Y$ are independent, conditioning on $X$ doesn't affect $Y$. Hence,

\begin{align}
    \mathbb{E}(Y \mid X> \frac{\exp(a + b)}{a^2 + b^2}) = \mathbb{E}(Y) = \int_0^{\infty} b y e^{-by} dy = \frac{1}{b}
\end{align}

- We have

\begin{align}
\mathbb{P}(0 < X < \sqrt{Y}) &= \int_0^\infty \int_{0}^{\sqrt{y}} f_{X,Y}(x,y) dx dy \\
    & = ab \int_0^\infty \int_{0}^{\sqrt{y}} e^{-(ax + by)} dx dy \\
    & = \frac{a \sqrt{\pi} \cdot e^{a^2/4 b} \cdot \text{erfc}\left(\dfrac{a}{2 \sqrt{b}}\right)}{2 \sqrt{b}}
\end{align}

### Subproblem 1 [10 Points]

Consider a mixture of three univariate Gaussian distributions:
$p(x) = \alpha p_1(x) + \beta p_2(x) + \gamma p_3(x)$

where $\alpha, \beta, \gamma \in [0,1]$ are mixture weights satisfying $\alpha + \beta + \gamma = 1$ and $p_1(x)$, $p_2(x)$ and $p_3(x)$ are univariate Gaussians with different parameters $(\mu_1, \sigma_1) \neq (\mu_2, \sigma_2) \neq (\mu_3, \sigma_3)$.

Derive the expectation and variance of $p(x)$, analytically, using their definitions.

### Answer

By linearity of the expectation, we have that if $X \sim p$
\begin{align}
    \mu = \mathbb{E}[X] = \alpha\mu_1 + \beta \mu_2 + \gamma \mu_3
\end{align}

Now, for the variance, we have that
\begin{align}
    \sigma^2 = \mathrm{Var}[X] &= \mathbb{E}[(X - \mu)^2] \\
                               &= \mathbb{E}[X^2] - \mu^2 \\
                               &= (\alpha\mathbb{E}[X_1^2] + \beta \mathbb{E}[X_2^2] + \gamma \mathbb{E}[X_3^2]) - \mu^2 \\
                               &= (\alpha (\sigma_1^2 + \mu_1^2) + \beta (\sigma_2^2 + \mu_2^2) + \gamma (\sigma_3^2 + \mu_3^2)) - \mu^2
\end{align}

Note that above, we utilized that $\sigma_i^2 = \mathbb{E}[X_i^2] - \mu_i^2$

## Problem 4

Consider two multivariate normal distributions $p(x) \sim \mathcal{N}(\mathbf{x}; \mathbf{\mu}_1, \mathbf{\Sigma}_1)$ and $q(x) \sim \mathcal{N}(\mathbf{x}; \mathbf{\mu}_2, \mathbf{\Sigma}_2)$, with $\mathcal{N}(\mathbf{x}; \mathbf{\mu}, \mathbf{\Sigma})$ defined as

\begin{align}
    \mathcal{N}(\mathbf{x}; \mathbf{\mu}, \mathbf{\Sigma}) = \frac{1}{\sqrt{(2 \pi)^n |\mathbf{\Sigma}|}} \exp \left[ -\frac{1}{2} (\mathbf{x} - \mathbf{\mu})^\top \Sigma^{-1} (\mathbf{x} - \mathbf{\mu}) \right]
\end{align}

- Analytically find the KL divergence $\mathbb{KL}(P || Q)$
- Assume $\mu_1 = [0, 1], \mu_2 = [2, 1], \sigma_1 = [[1, 2];[2, 0.6]], \sigma_2 = [[0.7, 3];[3, 1.5]]$. Sample 1,000 data from each distribution and use kernel density estimation to fit a probability density to the generated samples. Compute the KL-divergence numerically and compare it with the result you obtained from question (1). If you increase / reduce the number of samples, what do you observe?

## Answer

\begin{align}
    \mathbb{KL}(p||q) & = \int_{-\infty}^{\infty} p(x) \cdot \ln\left(\dfrac{p(x)}{q(x)}\right)dx\\
    & = \mathbb{E}_p \left[\ln p(x)  -\ln q(x) \right]dx\\
    & = \dfrac{1}{2} \mathbb{E}_p \left[ \ln \left( \dfrac{|\Sigma_q|}{|\Sigma_p|} \right) - \dfrac{1}{2} (\mathbf{x} -\mathbf{\mu}_p)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_p) + \dfrac{1}{2} (\mathbf{x} -\mathbf{\mu}_q)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_q)\right] \\
    & = \dfrac{1}{2} \ln \left( \dfrac{|\Sigma_q|}{|\Sigma_p|} \right) - \dfrac{1}{2} \mathbb{E}_p \left[ (\mathbf{x} -\mathbf{\mu}_p)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_p)\right] + \dfrac{1}{2} \mathbb{E}_p \left[ (\mathbf{x} -\mathbf{\mu}_q)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_q)\right] \\
\end{align}
Note that the second term above is a scalar. Hence, we can utilize the trace trick to simplify the expression as

\begin{align}
    \mathbb{E}_p \left[(\mathbf{x} -\mathbf{\mu}_p)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_p) \right] 
    & = \mathbb{E}_p \left[ \text{tr} \left[ (\mathbf{x} -\mathbf{\mu}_p)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_p)\right]\right] \\
    & = \text{tr} \left[ \mathbb{E}_p\left[ (\mathbf{x} -\mathbf{\mu}_p) (\mathbf{x} - \mathbf{\mu}_p)^\top \right]\Sigma_p^{-1} \right]\\
    & = \text{tr} \left[ \Sigma_p \Sigma_p^{-1} \right]\\
    & = \text{tr} \left[ I \right]\\
    & = d
\end{align}
where $\mathbf{x} \in \mathbb{R}^d$. 
Similarly,
\begin{align}
    \mathbb{E}_p \left[(\mathbf{x} -\mathbf{\mu}_q)^\top \Sigma_q^{-1} (\mathbf{x} - \mathbf{\mu}_q) \right]
    & = (\mathbf{\mu}_p -\mathbf{\mu}_q)^\top \Sigma_q^{-1} (\mathbf{\mu}_p - \mathbf{\mu}_q) + \text{tr} \left[\mathbb{E}_p\left[(\mathbf{x} -\mathbf{\mu}_p)  (\mathbf{x} - \mathbf{\mu}_p)^\top\Sigma_q^{-1}\right]\right]\\
    & = (\mathbf{\mu}_p -\mathbf{\mu}_q)^\top \Sigma_q^{-1} (\mathbf{\mu}_p - \mathbf{\mu}_q) + \text{tr} \left[\mathbb{E}_p\left[(\mathbf{x} -\mathbf{\mu}_p)  (\mathbf{x} - \mathbf{\mu}_p)^\top\right]\Sigma_q^{-1}\right]\\
    & = (\mathbf{\mu}_p -\mathbf{\mu}_q)^\top \Sigma_q^{-1} (\mathbf{\mu}_p - \mathbf{\mu}_q) + \text{tr} \left[ \Sigma_p\Sigma_q^{-1}\right]\\
\end{align}

$$ 
\implies \mathbb{KL}(p||q) = \dfrac{1}{2} \left[ \ln \left( \dfrac{|\Sigma_q|}{|\Sigma_p|} \right) - d +  (\mathbf{\mu}_p -\mathbf{\mu}_q)^\top \Sigma_q^{-1} (\mathbf{\mu}_p - \mathbf{\mu}_q) + \text{tr} \left[ \Sigma_p\Sigma_q^{-1}\right]\right]
$$


```python
import jax.numpy as np
from jax import random, jit, vmap, grad, jacfwd
from jax.scipy.stats import multivariate_normal
from jax.tree_util import tree_multimap
from jax.experimental import optimizers
from scipy.integrate import nquad

import numpy as onp
from scipy.stats import gaussian_kde
import pylab as pl
from IPython.display import clear_output
```

```python
mu_1 = np.array([0, 1])
mu_2 = np.array([2, 1])

sigma_1 = np.array([[1.4, 0.8], [0.8, 1.2]])
sigma_2 = np.array([[0.7, 0.4], [0.4, 0.8]])
```

```python
def analytic_KL(mu1, mu2, Sigma1, Sigma2):
    k  = mu1.size
    det1 = np.linalg.det(Sigma1)
    det2 = np.linalg.det(Sigma2)

    KL = 0.5 * (np.log(det2 / det1) - k 
                + (mu1 - mu2).T @ np.linalg.inv(Sigma2) @ (mu1 - mu2) 
                + np.trace(np.linalg.inv(Sigma2) @ Sigma1)
               )
    return KL
```

```python
def return_X_dists_KL(num_samples):
    # Creating realizations for the distributions:
    rng_key     = random.PRNGKey(7)
    key1, key2  = random.split(rng_key)
    X1          = random.multivariate_normal(key1, mu_1, sigma_1, shape = (num_samples,))
    X2          = random.multivariate_normal(key2, mu_2, sigma_2, shape = (num_samples,))

    # Performing Kernel Density Estimation
    p = gaussian_kde(X1.T)
    q = gaussian_kde(X2.T)

    # Computing KL divergence:
    # Considering some tolerances in numerator and denominator for float32
    kl_div = nquad(lambda x, y: p([x, y]) * np.log((p([x, y]) + 1e-7) / (q([x, y]) + 1e-7)), [[-7, 7], [-7, 7]])[0]
    return X1, X2, p, q, kl_div

X1, X2, p, q, kl_div = return_X_dists_KL(1000)
print('Numerically Computed KL Divergence(1000 Samples):', kl_div)

# Now reducing / increasing number of samples:
_, _, _, _, kl_div1 = return_X_dists_KL(500)
_, _, _, _, kl_div2 = return_X_dists_KL(2000)

print('Numerically Computed KL Divergence(500 Samples):', kl_div1)
print('Numerically Computed KL Divergence(2000 Samples):', kl_div2)
print('Analytic KL Divergence:', analytic_KL(mu_1, mu_2, sigma_1, sigma_2))
```

In general, we should notice that an increase in the number of samples would lead to a better estimation. However, this may not always be observed in this code due to errors in the KDE / integrator.

```python
# Creating realizations for Z:
num_samples = 2000
rng_key     = random.PRNGKey(0)
Z           = random.normal(rng_key, (num_samples, 2))

# Defining the functions g and g_inv:
def g_inv(z):
    z1 = z[0]
    z2 = z[1]
    x1 = np.sign(z1) * abs(z1)**(1 / 3) + 1
    x2 = z2 - 0.5    
    return np.array([x1, x2])

def g(x, coeff_x1_basis, coeff_x2_basis):
    x1 = x[0]
    x2 = x[1]
    z1 = np.sum(x1**(np.arange(0, coeff_x1_basis.size)) * coeff_x1_basis)
    z2 = np.sum(x2**(np.arange(0, coeff_x2_basis.size)) * coeff_x2_basis)
    return np.array([z1, z2])

mu     = np.zeros(2)
Sigma  = np.eye(2)
logp_z = multivariate_normal.logpdf(Z, mean = mu, cov = Sigma)
```

```python
# Getting vmapped versions for g and g_inv:
g_inv_vec = vmap(g_inv, in_axes = (0,))
g_vec     = vmap(g, in_axes = (0, None, None))

# Generating pairs of X:
X = g_inv_vec(Z)
# Checking that the given forward and inverse transforms are consistent:
coeff1_true = np.array([-1, 3, -3, 1])
coeff2_true = np.array([0.5, 1])
assert(np.mean(abs(g_vec(X, coeff1_true, coeff2_true) - Z)) < 5e-7)
```

```python
J       = jacfwd(g)
# Getting vmapped versions of J and det:
J_vec   = vmap(J, in_axes = (0, None, None))
det_vec = vmap(np.linalg.det)
# logp_x  = (  multivariate_normal.logpdf(g_vec(X), mean = mu, cov = Sigma)
#            + np.log(np.abs(det_vec(J_vec(X)))))
```

```python
# Framing the gradient ascent problem as a gradient descent:
@jit
def loss(params, batch):
    coeff1, coeff2 = params
    return -np.mean(multivariate_normal.logpdf(g_vec(batch, coeff1, coeff2), 
                                               mean = mu, cov = Sigma) + 
                    np.log(np.abs(det_vec(J_vec(batch, coeff1, coeff2)))))
    
# Computing Gradients:
grad_loss = jit(grad(loss))
```

```python
# Running this block will show a visualization of progress for p(x)
# We assume polynomial basis P_3 for z1:
coeff1_est = random.normal(rng_key, (4,))
# We assume polynomial basis P_1 for z1:
coeff2_est = random.normal(rng_key, (2,))

params = coeff1_est, coeff2_est
batch  = X

# Gradient descent
# Learning Rate
alpha = 0.005
n_its = 10000

# Variables for Plotting:
x    = np.linspace(-5.0, 5.0, 401)
y    = np.linspace(-5.0, 5.0, 401)
X, Y = np.meshgrid(x, y)
grid = np.hstack([X.reshape(-1, 1), Y.reshape(-1, 1)])

loss_history = [loss(params, batch)]
frames = []
for i in range(n_its):
    g = grad_loss(params, batch)
    params = tree_multimap(lambda x, g: x - alpha * g, params, g)

    # Track progress:
    loss_history = loss_history + [loss(params, batch)]

    # Print Progress:
    if i % 100 == 0:
        clear_output(wait=True)
        print('Iteration:', i)
        print('Loss:', loss(params, batch))

        coeff1_est, coeff2_est = params
        Z = np.exp(  multivariate_normal.logpdf(g_vec(grid, coeff1_est, coeff2_est), 
                                                mean = mu, cov = Sigma) 
                    + np.log(np.abs(det_vec(J_vec(grid, coeff1_est, coeff1_est)))))

        Z = Z.reshape(X.shape)

        pl.figure(dpi = 150)
        pl.title("Iteration = {}, Loss = {}".format(i, loss(params, batch)))
        
        pl.hist2d(batch[:, 0], batch[:, 1], bins=50, range=[(-5,5), (-5,5)], density=True)
        pl.xlim(-5, 5)
        pl.ylim(-5, 5)        
        pl.contour(X, Y, Z, levels=10, lw=2)
        pl.show()
```
