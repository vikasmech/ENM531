```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y):
  return np.sin(2.0*np.pi*x)*np.cos(2.0*np.pi*y)
```

```python
dim = 2
nn = 50
xx = np.linspace(0, 1, nn)
yy = np.linspace(0, 1, nn)
grid = np.meshgrid(xx, yy)
X = np.array(grid).T.reshape(-1,dim)
Y = f(X[:,0], X[:,1])
Z = f(grid[0], grid[1])
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,1], c = Y)
plt.colorbar()
plt.subplot(1,2,2)
plt.pcolor(grid[0], grid[1], Z)
plt.colorbar()
plt.tight_layout()
```
